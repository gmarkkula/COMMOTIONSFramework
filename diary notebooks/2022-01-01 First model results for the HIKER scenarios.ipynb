And below the - quite interesting - model results:
* The two models with evidence accumulation but no perceptual filtering - `oVAoBEvoAIoEAoSNc` and `oVAoBEvoAIoEAoSNv` - essentially always cross late, i.e., essentially never cross in front of a moving vehicle. I have looked at this and it seems to occur because the noise levels are high enough that things look too scary until it's anyway too late to cross in front of the vehicle. **So with lower noise levels in the probabilistic fits these models might be able to behave more sensibly.**
* The two models without evidence accumulation - `oVAoBEvoAIoSNv` and `oVAoBEvoAIoSNvoPF` - essentially always cross early, i.e., essentially always cross in front of a moving vehicle. **Possibly higher noise levels in the probabilistic fits might help these models.** But I am less hopeful about this - I think the early crossing to a large extent happens because there is no evidence accumulation to "protect" against purely noise-driven movement onsets.
* The model with both evidence accumulation and perceptual filtering - `oVAoBEvoAIoEAoSNvoPF` - exhibits both early and late crossing, and is qualitatively much more similar to the empirically empirically observed behaviour. This looks quite promising I think!
* The two models with perceptual filtering - `oVAoBEvoAIoSNvoPF` and `oVAoBEvoAIoEAoSNvoPF` - show a non-negligible frequency of negative CITs. I believe this is due to the implementation of the perceptual filtering, which early on in the simulation can result in sensory samples which say that the vehicle has already past the pedestrian, while it is actually still quite far away. **I think I should consider modifying the emulation of the "first HIKER vehicle" to instead add a large negative value for being within safety distance of the conflict space before the time at which the (emulated) first vehicle passes.**
