Rerunning the same fits as yesterday, but with the expanded parameter grid.

Overall, the results change quite a bit with the width/length functionality added (and possibly also due to that bug fix for the non-`oVA` model). Overall, in line with what I wrote yesterday, I think the differences are due to the change in the kinematic nature of the interaction - after the width/length modification, the pedestrian now has a considerably smaller conflict space to pass, and the car a slightly larger one, and this seems to change things up. For example, a lot of the `oVA` models now achieve "pedestrian start before vehicle stop", but they do so by not hesitating and instead speeding up to pass in front of the vehicle - which has now become feasible/sensible as the scenario is defined (see example simulation at the end of this notebook). This type of behaviour is not what "pedestrian start before vehicle stop" is aimed at identifying. I am thinking that what I should do next is:

* Break apart the `ActPed*` scenarios into two parts, to study the hesitation separately and the early start/speedup separately. This is anyway the sensible thing to do for the probabilistic fits, since noise will mean that the relevant kinematical conditions for the second part of these scenarios is created even more rarely than in the deterministic runs.
* Increase the 0.5 s and 0.5 m safety margins to 1 s and 1 m, now that the dimensions of the pedestrian has been reduced to more sensible values.
