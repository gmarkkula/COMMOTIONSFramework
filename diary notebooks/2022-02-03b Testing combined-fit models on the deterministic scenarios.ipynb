After the tweaks earlier today to the model/parameterisation retention criteria, I was curious to see how the combined-fit models could be expected to do on the full set of scenarios used in the deterministic fit. So I created a simple test for this with the `oVAoBEvoAIoEAoSNvoPF` model, not based on applying the criteria for actual combined fits, but just drawing 50 parameter combinations from the deterministic fits (those parameterisations of `oVAoBEvoAI` achieving all four main deterministic criteria) and the probabilistic fits (those parameterisations of `oVAoEAoSNvoPF` achieving all five probabilistic criteria).

The results below look quite nice. 
* A slight hint of a tendency to still do priority assertion (from about 2 s), but more noticeably a lot of hesitation from the car, which is perhaps reasonable with a pedestrian standing close to the kerb. 
* Also still quite a bit of pedestrians remaining stuck in the `PedCrossVehYield` scenario. 

It will be interesting to see how much the above changes after doing proper combined fits, and also possibly adding in `oBEo` with "confident" parameterisations could help alleviate some of the hesitation.
