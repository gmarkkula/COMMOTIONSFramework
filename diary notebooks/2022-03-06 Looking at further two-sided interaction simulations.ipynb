## More detailed looks at the two-sided interaction scenarios I have used since before

Running retained parameterisations one of the probabilistic models (`oVAoEAoSNvoPF`) and one of the combined models (`oVAoBEvoAIoEAoSNvoPF`), 50 randomly drawn parameterisations each for now, across the three interactive (two-agent) scenarios used in the probabilistic and combined fits, but running them for up to 20 s now to get a more complete view. Excluding parameterisations that get stuck (don't enter the conflict space) in one or more scenarios. Some things to note below:
* Clear signs of short-stopping in `EncounterPedPrio`, and some of priority assertion in `Encounter` (although a lot of car hesitation there also). 
* Clear advantage for the vehicle in the encounter (cf. Varhelyi, 1998).
* The model with `oBEvoAI` yields more efficient interactions than the simpler model in the two encounter scenarios, less clear effect in the pedestrian lead scenario.
* I am looking also at PETs, comparing to a digitized version of Fig. 13 from (Zhu et al., 2021) for reference. From the PETs I take the following:
    * The PETs one gets depend a lot on the initial kinematic situation. So trying to reproduce a PET curve without knowing those initial kinematic situations (such as for me with the Zhu et al paper now) may not be meaningful.
    * The PETs for the encounter scenarios could possibly be taken to suggest that my model is a bit more aggressive than the road users in the Zhu et al dataset - but again maybe there weren't many pure encounter scenarios in that dataset.
    * Overall I guess it could be argued that the results for the model are "compatible" with the Zhu et al dataset, but firmer conclusions than that don't seem possible.
