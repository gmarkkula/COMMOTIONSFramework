## Priority assertion
* The base (non-affordance) model can't do it.
* The `oVAa` model can do it, but only with the addition of both `oVAl` and `oBEvoAI`
* The `oVA` model can do it with `oBEvoAI` alone
* When `oVAl` is added to `oVA`, any `oBE*` addition is enough.

In sum, the simplest model that can do priority assertion is `oVAoBEvoAI`, and other `oBE*` combinations can only do it if combined with `oVAl`.

## Short-stopping
* From the figure above - with respect to short-stopping in the sense of showing larger early acceleration than strictly needed for yielding:
    * Most models can do it.
* From the figure further below (analysing the results of the follow-up short-stopping check, only of those model parameterisations meeting the early deceleration criterion above) - with respect to short-stopping in the sense of showing a large final stopping distance (the threshold in the run below is set to the vehicle's length of 4.2 m, base on Risto et al:
    * The base (non-affordance) model, the `oVAa`, and `oVAaoBEc` can't do it, even if adding `oBEv/o` assumptions.
    * Adding `oBEv` to `oVA` makes the model capable of doing it, even more so if also adding `oAI`.
    * If adding `oVAl` to `oVA`, the model can to some extent do it with just `oBEo`, even without `oBEv`. 
    * If adding `oVAl` to `oVAa`, the model can do it if also adding any `oBE*` assumption, especially easily for `oBEvoAI`.
    
In sum, the simplest model that can do short-stopping is `oVAoBEv`, but adding `oAI` makes the short-stopping more pronounced. Like for priority assertion, adding `oVAl` makes some additional models capable of this behaviour, for various `oBE*` assumptions.

## Pedestrian hesitation in deceleration scenario
All models can do this.

## Early pedestrian crossing
* The base (non-affordance) model can do this quite easily.
* The affordance-based models can't do this in their simplest form, but they can with the addition of any `oBE*`.

## Pedestrian hesitation in constant-speed scenario
* Without any `oBE*` assumptions, none of the models can do it.
* Adding `oBEo` makes any model capable of doing it, but as we have seen in the above results, only for model parameterisations which don't meet at least two of the other criteria.
* For the non-affordance-based model and for `oVAaoVAl`, there are also other `oBE*` combinations besides pure `oBEo` that are capable of doing it, but again only for parameterisations which don't meet at least two of the other criteria.

In sum, there are ways of getting the deterministic models to do pedestrian hesitation in the constant-speed scenario, especially using the `oBEo` assumption, but not while also achieving the other phenomena.
