Note that this commit is on an unmerged branch - see further below.

While an ARC4 run was going on, I was considering some changes to the deterministic metrics/simulations:
1. I was thinking that it would be nicer to reformulate the short-stopping metric in terms of actual stopping margin, defined as the vehicle margin to the conflict space being larger than 2 m when the vehicle speed goes below 0.5 m/s (some of the models can stay just above zero speed for quite a while without stopping completely).
2. I happened to notice that I was using `metric_ped_av_speed` for the deterministic simulations of the `PedHesitate*` scenarios, but `metric_ped_av_speed_to_CS` for the probabilistic `PedHesitateVehConst` scenario; I changed to the latter when I realised that for some models the "hesitation deceleration" comes after the halfway point to the conflict space which `metric_ped_av_speed`. To be consistent I have now changed to using `metric_ped_av_speed_to_CS` also for the deterministic simulations (I don't expect this to make a difference to the deterministic results).
3. To enable both of the changes above, I needed to extend the maximum simulation end time from the 8 s I was previously using for all deterministic simulations, to 12 s, which is what I have been using for the probabilistic simulations. For simplicity, I changed it to 12 s for all simulations, but this in itself shouldn't affect run times since many of the simulations which would otherwise be left at 8 s will anyway be terminated early due to the various stopping criteria I have implemented.

Comparing the results below especially point 1 above does make some difference - the pure `oVA` model does not achieve short-stopping by this definition, and some models which didn't achieve it with the previous definition (or only for few parameterisations) do so now (or for more parameterisations). There are also some smaller changes in exactly what percentages of parameterisations achieve the hesitation criteria, but no qualitative differences.

While running this I have noticed though that the runs are taking about twice as long as previously, or maybe even a bit more. I think this is mainly due to points 1 and 2 above (extending simulation time from about 2 s to 12 s for the `VehShortStop` scenario, and from about 1 s to 2 s for the `PedHesitate*` scenarios, respectively). 

For this reason I will - at least for now - content myself with the previous "exaggerated early deceleration" definition of short-stopping, and I will leave this branch unmerged.
