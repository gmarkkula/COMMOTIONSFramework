Below I am also recreating Figs S17, S19, and S20 from the preprint for `oVAoBEvoAIoDAoSNvoPF`. Takeaways:
* The retention of parameterisations in Fig S17 is pretty similar to that for the `oEA` model. As noted in the preprint, the main driver for non-retention of parameterisations of the `oEA` model was the gap acceptance hesitation criterion (the requirement of hesitation in 80% of trials turns out be higher than what the model generally does) - I haven't checked now that the same is true for the `oDA` model, but given that the CDFs are so similar between the two models in the 2022-07-24 diary notes, that seems almost certain to be the case.
* The parameter exclusion for non-crossing in the HIKER simulations (Fig S19) is considerably reduced for the `oDA` model compared to the `oEA` model. And the same is true for the DSS simulations (Fig S20). Basically if $\xi_\mathrm{th}$ is low enough, the model doesn't get stuck. As we saw above, it may instead be more prone to collision however - that is something that could be looked into further.

Overall my takeaway is that `oDA` is an improvement over `oEA`, since it seems to have regions in parameter space which both achieve the targeted phenomena and basically never get stuck. It would be good, however, to look closer at what parameterisations have a tendency for collisions (I am guessing low $\xi_\mathrm{th}$ combined with high $\sigma_\mathrm{v}$).
