There are some interesting new types of predictions from these models. I think the overall takeaway is still reasonably that only models with both evidence accumulation and Bayesian perceptual filtering do a good job of predicting the patterns in the empirical data, but to make that case one needs to take a more complete look at "the patterns in the empirical data". Previously, it was enough to note the bimodal nature of the empirical CITs, and before `oDA` only models with both `oEA` and `oPF` predicted that bimodality. Now, with `oDA`, all or nearly all `oDAoSN*` do show (1) bimodality, but are missing out on one or several of: (2) the clear effect of time gap of the probability mass in the early-crossing mode (a possibility here is that some model parameterisations always cross early and some always late), (3) the overlap between constant speed distribution and the early mode of the yielding distribution, and/or (4) the timing of the late mode occurring quite late (especially the `oDAoSNc` models seem to put that mode too early).

The `oVAoBEvoAIoDAoSNvoPF` model variant is ticking all of these boxes, just like its analogue `oVAoBEvoAIoEAoSNvoPF` (thus far identified as the maximally successful model variant) does. The `oDA` variant arguably provides a qualitatively even higher similarity to the empirical CITs. And the fraction of non-crossing is much lower; only 22 % non-crossing for the `oDA` variant compared to 66 % for the `oEA` variant.

Still no model seems to clearly predict the impacts of vehicle speed in the empirical CITs. (For CIT < 5 s for time gaps 2-3 s, the empirical data seem to suggest a greater tendency to cross in front of slower vehicles. This aligns with Athanasios's and Kai's findings. In the 4-5 s time gaps, the familiar distance(/speed) effect, instead with a greater tendency to cross in front of faster vehicles, is clear in the empirical data.)

But overall really nice results. Next up: two-agent simulations, including DSS predictions.
