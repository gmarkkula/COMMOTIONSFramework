I have now added, also for the probabilistic and combined fits, the same kind of criterion threshold sensitivity analysis plots that I previously set up for the deterministic fits (2022-01-22 diary notes).

Some conclusions from the results for the probabilistic fits below:

* Collisions:
    * Adding `oEA` increases the tendency for collisions, not surprisingly.
    * Adding `oPF` does not seem to have consistent effect on collisions - it seems to increase the tendency for some model/scenario combinations, and decrease it for others. Some of this might be due to low numbers here, but I don't think I am just looking at noise here. Consider for example the apparent increase in collisions when adding `oPF` to the `oVAoEAoSNc` model, but the apparent decrease when adding it to the `oVAoEAoSNv` model. Both of these models were tested over 1000 parameterisations each, and the same pattern is there in both of the encounter scenarios.
    * Overall, `oSNc` models are more collision-prone than `oSNv` models (because they are less noisy closer to potential collisions) and than `oAN` models (because when collision risks arise, the involved values tend to be larger than the value noise amplitude).
    * Except for the `oVAoSNcoPF` and `oVAoEAoSNcoPF` models, the models have relatively few parameterisations which show collisions.
* Pedestrian hesitation:
    * As shown previously (2022-01-17b diary notes), the `oVA*oAN` models exhibit slowdown, but not relative to their natural free speed.
    * The only other model completely incapable of hesitation is `oVAoSNc`.
    * Generally it is easier to get the `oSNv` models to exhibit hesitation than it is with the `oSNc` models.
    * Adding `oPF` increases the tendency for hesitation.
    * Adding `oEA` decreases the tendency for hesitation.

Overall, the `oSNv` models have some properties that could arguably make them seem preferable to the `oSNc` models - less likely to collide and more likely to exhibit hesitation - but in general both types of models are capable of achieving the phenomena in question. Adding `oEA` to a model tends to decrease the tendency for hesitation, and increase the tendency for colliding. Adding `oPF` tends to increase the tendency for hesitation, but with a less clear effect on the tendency for colliding.

With regards to the combined fits further below, much of the same general observations seem to hold. I haven't gone through this in massive detail now, but I have noted for example that the models with greatest tendency for collisions include `oVAa` and/or `oSNc`.
