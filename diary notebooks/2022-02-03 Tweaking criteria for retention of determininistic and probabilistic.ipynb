## Retention of deterministic parameterisations

Lower the prio assertion threshold to 1.005 to acknowledge that `oVAoBEvoAI` does achieve prio assertion, and retain models with at least one parameterisation which achieves all four main criteria (as previously defined), and for these models retain parameterisations which:
1. Achieve at least three of the four main criteria (as previously), but which also:
2. Don't get stuck in any of the kinematic variants of PedCrossVehYield.

It turns out that the second point above does not make a huge difference - only moves the percentage of included parameterisation down by about 4 points for most models, even less for others (not seen below, verified in separate tests), but I'll keep it anyway, because it clearly serves to exclude parameterisations which are too uncertain in their behaviour estimation, to the point that they freeze. It is interesting to see that the tendency to freeze like this is almost entirely driven by the `oBE*` assumptions, with very little difference between `oVA` and `oVAa` and whether `oVAl` is included or not.

Looking at the retained parameters more in general, the most noticeable difference compared to the most recent retention (diary notes 2022-01-12) is that a chunk of high $\sigma_O$, low $\beta_V$ parameterisations no longer get included. Otherwise the retained parameters look quite similar to before, just generally extending into slightly lower values of $\beta_V$, presumably due to the lowered threshold for prio assertion.

## Retention of probabilistic parameterisations

The way I do it now is clearer I think. Looking at the retained parameterisations below they are very similar to the ones in the 2022-01-17b notes (not sure why they are not completely identical, not looking closer at this now), and the main difference compared to the parameterisations I retained in the 2022-01-19b notes is that I am now limiting $\Delta V_{th,rel}$ upward, just as I expected - too high values for this parameter can make the agents freeze.
