As noted in the 2021-11-12 notes, the `oAN` model variants have rather erratic-looking behaviour, and it's quite hard to understand what they're up to. But it seems quite clear that they are exhibiting hesitation, and seemingly even more so than the `oSNv` models.

The `oSNc` models are the most collision-prone.
