## In-depth looks / discussion

So now we find two models which are able to meet all four criteria: `oVAoBEvoAI` and `oVAoBEooBEvoAI`. As hinted at the top above, the reason this works now is because of the tweak of the `ActVehStatPed` scenario, and because of the weakening of the requirement for the "pedestrian start before vehicle stop" criterion.

Interestingly, none of the `oVAa` models managed to meet all four criteria. In particular they did not manage to meet the "asserting priority" criterion; generating an acceleration just before/at the time of reaching the conflict space rather than early on. I am not entirely sure why this is the case, but looking closely at these simulations there is actually a small acceleration already from the start, which does seem to be due to the car wanting to assert priority, and my guess is that since these accelerations get extrapolated into the future, the car judges that it has now asserted its priority, so further acceleration is not needed. Until just before entering the conflict space apparently - this part is even more puzzling. But I am happy to discard the `oVAa` formulations based on these results - they feel more difficult to make sense of, and the beliefs they generate are also pretty messy-looking.

Looking closer at the parameter values for the successful `oVAoBEvoAI` and `oVAoBEooBEvoAI` models below:
* There is a redundancy, whereby $T_\delta \leq 30$ s goes with $\beta_V = 15$, and $T_\delta = 60$ s goes with $\beta_V = 45$
* All combinations of $T_{Of}$ and $\sigma_O$ work, except $\sigma_O = 0.02$ m.

```python
for model in ('oVAoBEvoAI', 'oVAoBEooBEvoAI'):
    print(f'Parameterisations for model {model} meeting all four criteria:')
    fit = det_fits[model]
    i_good_parameterisations = np.nonzero(fit.n_criteria_met == 4)[0]
    for i_param in i_good_parameterisations:
        print(fit.get_params_dict(fit.results.params_matrix[i_param, :]))
    print()
```
