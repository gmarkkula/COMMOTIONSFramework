## In-depth looks / discussion

Basically the non-affordance-based models aren't able to do much with any of their extra assumptions. Already the base model can actually do "hesitation + start before vehicle stop" in the yielding scenario, because the model reacts to the low apparent initial TTA, but then with a sufficiently low $k_c$ it can still get started before the vehicle has fully stopped. Note that $k_c = 0.43$ for the parameterisation highlighted for the base model above - a lower $k_c$ than needed to ensure collision-free simulations in my diary notes from earlier today. 

The models with `oAI` does show some priority assertion, but only for parameterisations which do not achieve the "start before vehicle stop" criterion. These models also show a slight tendency toward short-stopping, but not enough to register as meeting the criterion as I have defined it.

```python

for model in ('oBEvoAI', 'oBEooBEvoAI'):
    print(model)
    fit = det_fits[model]
    i_crit = MAIN_CRITERIA.index('veh_assert_prio')
    met_crit = np.nonzero(fit.main_criteria_matrix[i_crit,:])[0]
    i_parameterisation = met_crit[0]
    free_params = fit.results.params_matrix[i_parameterisation,:]
    print(fit.get_params_dict(free_params))
    fit.set_params(free_params)
    crit_array = fit.main_criteria_matrix[:, i_parameterisation]
    print('Main criteria met: ', end='')
    print(crit_array)
    for scenario in sc_fitting.DET1S_SCENARIOS.values():
        print(scenario.name)
        sim = fit.simulate_scenario(scenario)
        sim.do_plots(kinem_states=True, beh_probs=True)
```
