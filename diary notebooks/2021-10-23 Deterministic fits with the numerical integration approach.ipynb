### Just unpacking the short-stopping a bit

I haven't looked in much detail at the short-stopping before, but I have noticed that also the base `oVA` model has shown some tendency of short-stopping, but not enough to meet the current criterion of exceeding the required stopping deceleration by 2 m/s^2 or more. Below it can be seen that the tendency to increase deceleration for this model comes from a judgment that this will make the rest of the deceleration to pass 2nd less uncomfortable; in the snapshot the value for the 1 m/s^2 deceleration is the lowest at -0.7, but the remaining value for achieving the "pass 2nd" access order is the least negative at -4.6. This is the reason why this model ends up doing quite a bit of short-stopping.

In the second simulation below, of the `oVAoBEvoAI` model, it can be seen that the reason that this model is doing even shorter stopping is because it judges that the value for the pedestrian of passing in second will be reduced further by the car decelerating further. So interestingly, here it is not about increasing the value for the pedestrian of passing first - that value is constant across all vehicle actions in the shown snapshot, late in the situation. (However, looking at some other snapshots earlier in the simulation, there seems to be a bit of effect on the value for the pedestrian of passing first there, but quite small effects.)

```python
for model in ('oVA', 'oVAoBEvoAI'):
    print(f'*** Model {model}')
    fit = det_fits[model]
    if model == 'oVA':
        free_params = (15,)
        snapshot_time = 0.3
    else:
        free_params = (60, 45)
        snapshot_time = 4.7
    print(fit.get_params_dict(free_params))
    fit.set_params(free_params)
    print(f'Snapshot at t = {snapshot_time} s:')
    sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActVehStatPedPrio'], snapshots=(None, (snapshot_time,)))
    print(f'Car stopped at x = {sim.agents[sc_fitting.i_VEH_AGENT].trajectory.pos[0,-1]:.1f} m')
    sim.do_plots(kinem_states=True)
    print('')
```
