### Asserting priority - no models are doing it, because the scenario is not quite right

I was a little surprised to see that the models with `oAI` were not having the car asserting priority in the `ActVehStatPed` scenario - this scenario seemed  quite similar to the "baseline scenario" where I have seen the model assert priority before (e.g., the 2021-09-16b diary notes). The difference, though, is that in the `ActVehStatPed` the pedestrian is stationary near the crossing, rather than approaching with a similar TTCA to the vehicle. So in this scenario, with TTCA_0 = 4 s as I have simulated it above, the car judges that the most sensible thing for the pedestrian to do is to pass first. If I dial down the TTCA_0 to 2 s, where the car thinks it has a slight edge, we now get an assertive speed increase, especially if I also dial up $\beta_V$ (the last three simulations below):

```python
fit = det_fits['oVAoBEvoAI']
def assertiveness_tests(T_P):
    for beta_V in (5, 15):
        free_params = [T_P, 30, beta_V]
        print(fit.get_params_dict(free_params))
        fit.set_params(free_params)
        for ttca0 in (4, 3, 2):
            print(f'Vehicle TTCA_0 = {ttca0} s:')
            scenario = sc_fitting.SCPaperScenario('ActVehStatPedVariation',
                                                  initial_ttcas=(math.nan, ttca0),
                                                  ped_start_standing=True,
                                                  ped_const_speed=True)
            sim = fit.simulate_scenario(scenario)
            sim.do_plots(kinem_states=True, beh_probs=True)
assertiveness_tests(T_P=0.4)
```

### Short-stopping

#### All models do a little bit of short-stopping, some do more
It's not so easy to define exactly what short-stopping is, unless you simulate all the way to a full stop. Currently I define it as short-stopping if the deceleration is ever at least 3 m/s^2 more than it needs to be to stop exactly at the conflict space. With this definition, all models show some of it - not only models with `oBEvoAI` as I expected. I was surprised to see the pure `oVA` model do short-stopping - it turns out it does so in the $T_P = 0.8$ s case. Here is what I am used to seeing, for a lower $T_P$ value:

```python
fit = det_fits['oVA']
for T_P in (0.4, 0.8):
    free_params = [T_P, 30]
    print(fit.get_params_dict(free_params))
    fit.set_params(free_params)
    sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActVehStatPedPrio'])
    sim.do_plots(kinem_states=True)
```

In the models with `oAI`, a much larger fraction of the parameterisations do short-stopping. This is the one shown above for `oVAoBEvoAI`, which is doing short-stopping also for $T_P = 0.4$ s. And when cranking up $\beta_V$, there is an increase in the short-stopping tendency:

```python
fit = det_fits['oVAoBEvoAI']
for beta_V in (5, 15, 45):
    free_params = [0.4, 30, beta_V]
    print(fit.get_params_dict(free_params))
    fit.set_params(free_params)
    sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActVehStatPedPrio'])
    sim.do_plots(kinem_states=True, beh_probs=True)
```

#### Not clear whether these car yielding profiles are very human-like...?
Most of the model variants tend to decelerate in the yielding scenario the way they accelerate from standstill, with an early peak that falls down to zero (see the `oVA` simulations above). Looking at human data for stopping at intersections (Da Lio et al, 2018), it looks pretty much the other way around, with a slow ramp up of deceleration to a peak just before stopping. The 'oVAoBEvoAI' simulations look a little bit nicer in that respect, but perhaps a bit too jumpy on the pedal. Ideally we should compare to human acceleration traces when yielding to pedestrians though... 

When looking back at the earliest stopping tests with the model, in `acceleration ctrl tests/`, there is more of a an early ramp-up.

### Pedestrian hesitation and speed-up in constant-speed scenario

#### Only really seeing it with pure `oBEo` model

Some nice behaviour from the `oVAoBEo` model in the `ActPedLeading` scenario; it is uncertain about whether the car intends to pass first or second, and thus slows down, and with the model parameterisation identified above it stays at that lower speed long enough to need to speed up to higher than free speed while crossing. I was concerned about whether this behaviour was very dependent on the quick forgetting of behaviour observation evidence ($T_Of = 0.5$ s), or the long prediction time (with $T_P = 0.8$ s), but the slowing down in itself is robust to changing these parameters, as can be seen below:

```python
fit = det_fits['oVAoBEo']
for T_P in (0.8, 0.4):
    for T_Of in (0.5, 2):
        free_params = [T_P, 30, T_Of, 0.5]
        print(fit.get_params_dict(free_params))
        fit.set_params(free_params)
        sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActPedLeading'])
        print(f'Pedestrian min/max speed: ({np.min(sim.agents[sc_fitting.i_PED_AGENT].trajectory.long_speed)}'
              f', {np.max(sim.agents[sc_fitting.i_PED_AGENT].trajectory.long_speed)}) m/s')
        sim.do_plots(kinem_states=True, beh_probs=True)
```

To also see the speeding-up part with lower prediction time and slower forgetting, the scenario needs tweaking. Slightly odd-looking behaviour though, with an extra pulse of deceleration just as the pedestrian is entering the conflict space, but then it catches up with an acceleration:

```python
free_params = [0.4, 30, 2, 0.5]
print(fit.get_params_dict(free_params))
fit.set_params(free_params)
scenario = sc_fitting.SCPaperScenario('ActPedLeadingVariant',
                           initial_ttcas=(3, 6.5),
                           veh_const_speed=True)
sim = fit.simulate_scenario(scenario)
print(f'Pedestrian min/max speed: ({np.min(sim.agents[sc_fitting.i_PED_AGENT].trajectory.long_speed)}'
      f', {np.max(sim.agents[sc_fitting.i_PED_AGENT].trajectory.long_speed)}) m/s')
sim.do_plots(kinem_states=True, beh_probs=True, action_vals=True)
```

#### Difficult to sensibly get hesitation when adding `oBEv` to `oBEo`

I would have hoped that there might be a parameterisation of a model with `oBEooBEv` which could also show the hesitation above. In the simulations below I show first the same parameterisation as was highlighted in the overview above (with $\beta_V = 5$ and $\sigma_O = 0.02$ m), and then I show that I need to decrease $\beta_V$ to the very low value 0.1 and increase $\sigma_O$ to 0.5 m (as in the previous `oBEo` simulations above), to get the hesitation back. With the higher $\beta_V$, it just seems too obvious to the pedestrian that the car shouldn't pass first. And a similar story with the lower $\sigma_O$ - since the car's behaviour is more consistent with passing in second, given the pedestrian's initial speed. 

Overall I am thinking that this type of hesitation in pedestrians probably has more to do with initial perceptual uncertainty about where in the world the car is, rather than uncertainty about the car driver's intentions - so I might need to go to Kalman-like perception to really sensibly capture this phenomenon...

```python
fit = det_fits['oVAoBEooBEv']
for sigma_O in (0.02, 0.5):
    for beta_V in (5, 0.1):
        free_params = [0.8, 30, beta_V, 0.5, sigma_O]
        print(fit.get_params_dict(free_params))
        fit.set_params(free_params)
        sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActPedLeading'])
        print(f'Pedestrian min/max speed: ({np.min(sim.agents[sc_fitting.i_PED_AGENT].trajectory.long_speed)}'
              f', {np.max(sim.agents[sc_fitting.i_PED_AGENT].trajectory.long_speed)}) m/s')
        sim.do_plots(kinem_states=True, beh_probs=True)
```

### All models hesitate in the yielding scenario, but no models start crossing before the vehicle is at a full stop

It was quite expected that all models would have the pedestrian slow down in front of the decelerating vehicle. I was hoping though that there would be some parameterisation of at least the `oBEo` model which would show early crossing onset. 

The `oBEo` parameterisation highlighted in the overview above looks quite strange for this scenario, but it turns out this is because of the combination of quick forgetting ($T_{Of} = 0.5$ s) and especially the quite noisy behaviour observation ($\sigma_O = 0.5$ m). Since the acceleration need for passing first or second aren't massively different in this scenario in the later stages of the vehicle's approach, and also since the actual behaviour of the car - to stop a bit before the conflict space - also isn't exactly what is assumed for the "pass 2nd" behaviour, it seems the high $\sigma_O$ causes both behaviours to be about equally (un)likely, so the belief quickly fizzles out to 50/50:

```python
fit = det_fits['oVAoBEo']
free_params = [0.8, 30, 0.5, 0.5]
print(fit.get_params_dict(free_params))
fit.set_params(free_params)
sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActPedPrioEncounter'])
sim.do_plots(kinem_states=True, beh_probs=True, beh_accs=True)
```

Below I show first the model parameterisation that was highlighted above, and then the following simulations show that by making the observation less noisy ($\sigma_O = 0.1$ m) and/or slowing down the forgetting ($T_{Of} = 2$ s), we get more robust development of belief.

```python
for T_Of in (0.5, 2):
    for sigma_O in (0.5, 0.1):
        free_params = [0.8, 30, T_Of, sigma_O]
        print(fit.get_params_dict(free_params))
        fit.set_params(free_params)
        sim = fit.simulate_scenario(sc_fitting.DET1S_SCENARIOS['ActPedPrioEncounter'])
        sim.do_plots(kinem_states=True, beh_probs=True)
```

However, still no sign of early crossing onset, before the vehicle has come to a full stop, not even with the $T_P = 0.8$ s. This was sort of expected, given the results from for example 2021-08-27, but I had some hope that there might be some parameterisation of the `oBEo` or maybe `oBEooBEvoAI` models which could still achieve this. Instead it seems like the `oVAa` route may be the one to take, but if so it seems likely that this needs addition of a new value function term for (non-acceleration-aware) looming aversion.

* Potential for narrowing down the search space:
    * The $T_P = 0.8$ s simulations gave quite weird-looking results in both `ActVehStatPed` and `ActVehStatPedPrio`. (Is it an acceleration control thing?) Maybe we can stick to just $T_P = 0.4$ s for now...?
    * Based on the `ActPedPrioEncounter`, $T_{Of} = 0.5$ s seems too small - so maybe skip it from the searches...?
* Model / model fitting changes:
    * Abandon the idea of seeing pedestrian hesitation etc in the `ActPedLeading` scenario, until adding in some Kalman-like filtering.
    * Switch over to `oVAa` instead, in the first instance without changing anything else in the model. Based on previous tests this will require changing the metrics/criteria a little bit to ensure crossing is in the last seconds of the vehicle deceleration in `ActPedPrioEncounter`.
