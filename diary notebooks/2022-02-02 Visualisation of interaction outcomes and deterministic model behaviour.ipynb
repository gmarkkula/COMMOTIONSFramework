### Probabilistic fits

In the figure below, blue is for the pedestrian, green is the vehicle. The stacked bar is showing first-passer, and orange indicates neither agent exited the conflict space within the simulation time. The histograms are distributions of exit times. * Clear advantage for the car in encounters, across all models.
* In the ped priority scenario:
